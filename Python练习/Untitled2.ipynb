```python
# -!- coding: utf-8 -!-
import requests
import os
import time

def get_page(offset):
    '''访问页面，并且获取 json 格式的数据'''
    url = 'https://www.toutiao.com/search_content/?'
    #url请求传递的参数
    params = {
        'offset':offset,
        'format':'json',
        'keyword':'街拍',
        'autoload':'true',
        'count':'20',
        'cur_tab':'3',
        'from':'gallery',
    }
    #添加头信息
    header = {
        'authority':'www.toutiao.com',
        'method':'GET',
        'path':'/search_content/?offset=20&format=json&keyword=%E8%A1%97%E6%8B%8D&autoload=true&count=20&cur_tab=3&from=gallery',
        'scheme':'https',
        'accept':'application/json, text/javascript',
        'accept-encoding':'gzip, deflate, br',
        'accept-language':'zh-CN,zh;q=0.8',
        'content-type':'application/x-www-form-urlencoded',
        #cookie:UM_distinctid=15f1c13ed94905-013ba6720116af-3e63430c-144000-15f1c13ed95811; uuid="w:cd08eba15729466b8006b1b808652009"; tt_webid=6476830383903573518; WEATHER_CITY=%E5%8C%97%E4%BA%AC; __tasessionId=bvcdrux911518150992381; CNZZDATA1259612802=378713649-1508002287-https%253A%252F%252Fwww.baidu.com%252F%7C1518150741; tt_webid=6476830383903573518
        'referer':'https://www.toutiao.com/search/?keyword=%E8%A1%97%E6%8B%8D',
        'user-agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36',
        'x-requested-with':'XMLHttpRequest',
    }
    #访问url
    r = requests.get(url,params=params,headers=header)
    return r.json()

def get_images(json):
    '''从返回的json文件当中，提取出有用的信息，以这样的形式来保存[{'title':'abc','image_url':[url_1,url_2]},{}]'''
    #print(json['data'])
    result = []
    items = json['data']
    for item in items:
        image = item['image_detail']
        image_url = []
        for i in image:
            image_url.append(i['url'])
        #print(image_url)
        #print(len(image_url))
        temp = {
            'title':item['title'],
            'image_url':image_url,
        }
        result.append(temp)
    return result


def save_images(result):
    '''以每一个图集的title为文件名，建一个个文件，然后从得到的图片的url中下载图片，并一一放置在相对应图集的问价夹当中'''
    for item in result:
        if not os.path.exists('{}/{}'.format('街拍美图', item['title'])):
            os.mkdir('{}/{}'.format('街拍美图', item['title']))
        #os.mkdir('{}/{}'.format('街拍美图', item['title']))
    for i in result:
        image_url = i['image_url']
        for url in image_url:
            r = requests.get(url)
            url_name = url.split('/')[-1]
            file_path = '{}/{}/{}.jpg'.format('街拍美图',i['title'],url_name)
            print(file_path)
            with open(file_path,'wb') as f:
                f.write(r.content)




if __name__ == '__main__':
    page = int(input('请输入需要爬取的页数'))
    for i in range(page):
        print('正在爬取第%d页' %(i+1))
        offset = i*20
        html = get_page(offset)
        time.sleep(1)
        result  = get_images(json=html)
        save_images(result)
```
